```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import spacy
from simplemma import text_lemmatizer
import numpy as np
import ast
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
from fuzzywuzzy import fuzz
import tensorflow as tf
import tensorflow_hub as hub
nltk.download('stopwords')
nltk.download('punkt')
```

```python
target_column = 'Konu'

# Stop words listesi
stop_words = set(stopwords.words('turkish'))
```

```python
def remove_stopwords(text):
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
    return ' '.join(filtered_text)
```

```python
new_column = 'temizlenmis_konu'

# Stop words'leri temizle ve yeni sütuna ekle
df[new_column] = df[target_column].apply(remove_stopwords)
```

```python
df.head()
```

```python
# stop word ayıklaması yapmayan metin temizleme fonk.
def clean_text(text):
    text = text.replace("Â", "a")
    text = text.replace("â", "a")
    text = text.replace("î", "i")
    text = text.replace("Î", "ı")
    text = text.replace("İ", "i")
    text = text.replace("I", "ı")
    text = text.replace(u"\u00A0", " ")
    text = text.replace("|", " ")

    text = re.sub(r"@[A-Za-z0-9]+", " ", text)
    text = re.sub(r"(.)\1+", r"\1\1", text)
    text = re.sub(r"https?:\/\/\S+", " ", text)
    text = re.sub(r"http?:\/\/\S+", " ", text)
    text = re.sub(r"\n", " ", text)
    text = re.sub(r"#(\w+)", " ", text)
    text = re.sub(r"^\x00-\x7F]+", " ", text)
    text = re.sub(r"[^A-Za-zâîığüşöçİĞÜŞÖÇ]+", " ", text)
    text = re.sub(r"((https://[^\s]+))", " ", text)
    text = " ".join(text.lower().strip().split())

    return " ".join(text_lemmatizer(text, lang="tr"))
```

```python
def clean_row(row):
    konu = clean_text(row["temizlenmis_konu"])
    return konu
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['temizlenmis_konu'])
```

```python
def konu_yaz3(konu):
    kullanici_konu_tokenized = nlp(konu.lower())
    kullanici_vektor = vectorizer.transform([konu.lower()])
    benzerlik_skorlari = linear_kernel(kullanici_vektor, X).flatten()

    benzerlik_film_indeksleri = list(enumerate(benzerlik_skorlari))

    sirali_benzerlikler = sorted(benzerlik_film_indeksleri, key=lambda x: x[1], reverse=True)

    en_yakin_filmler = []
    for i in range(min(1, len(sirali_benzerlikler))):
        indeks, benzerlik_skoru = sirali_benzerlikler[i]
        onerilen_film = df['Film Adı'][indeks]
        en_yakin_filmler.append((onerilen_film, benzerlik_skoru))

    # En yüksek benzerlik oranına sahip olan filmin indeksini bulma
    en_yuksek_benzerlik_indeks = sirali_benzerlikler[0][0]

    # En yüksek benzerlik oranına sahip olan filmin konusunu alma
    en_yuksek_benzerlik_film_konu = df['Konu'][en_yuksek_benzerlik_indeks]

    return en_yakin_filmler, en_yuksek_benzerlik_film_konu

# Örnek kullanım
kullanici_konu = "Azılı suçlularla tıkabasa dolu bir hapishanenin yöneticileri, cezaevindeki mahkumları birbirleriyle dövüşmeye zorlar"
en_yakin_filmler, en_yuksek_benzerlik_film_konu = konu_yaz3(kullanici_konu)

# En yüksek benzerlik oranına sahip olan filmi ve en yüksek filmin konusunu bastırma
print("En Yüksek Benzerlik Oranına Sahip Film ve En Yüksek Film Konusu:")
for film, benzerlik_orani in en_yakin_filmler:
    print(f"{film}: Benzerlik Oranı = {benzerlik_orani:.4f}")

print("\nEn Yüksek Benzerlik Oranına Sahip Film Konusu:")
print(en_yuksek_benzerlik_film_konu)
```

```python
user_input = input("Film konusu giriniz: ")

# Veri setindeki film konularını Universal Sentence Encoder ile temsil et
film_konulari_embeddings = embed(df["temizlenmis_konu"].tolist()).numpy()

# Kullanıcının girdiği cümleyi Universal Sentence Encoder ile temsil et
user_input_embedding = embed([user_input]).numpy()

# Cosine Similarity ile benzerlik ölçümü yap
similarities = np.dot(film_konulari_embeddings, user_input_embedding.T)
most_similar_index = np.argmax(similarities)
most_similar_film = df.iloc[most_similar_index]

# Sonuçları yazdır
print("En yüksek benzerlik oranı:", similarities[most_similar_index][0])
print("En yüksek benzerlik oranına sahip film:")
print("Film Adı:", most_similar_film["Film Adı"])
print("Film Konusu:", most_similar_film["Konu"])
```

```python
user_input = input("Film konusu giriniz: ")

# Fuzz modelini kullanarak benzerlik ölçümü yap
df["Similarity"] = df["temizlenmis_konu"].apply(lambda x: fuzz.token_set_ratio(x, user_input))

# En yüksek benzerlik oranına sahip filmi bul
most_similar_index = df["Similarity"].idxmax()
most_similar_film = df.loc[most_similar_index]

# En yüksek benzerlik oranını ve sonuçları yazdır
print("En yüksek benzerlik oranı:", most_similar_film["Similarity"])
print("En yüksek benzerlik oranına sahip film:")
print("Film Adı:", most_similar_film["Film Adı"])
print("Film Konusu:", most_similar_film["Konu"])
```
